```python
# Dependencies
from splinter import Browser
from bs4 import BeautifulSoup as bs
from webdriver_manager.chrome import ChromeDriverManager

executable_path = {'executable_path': ChromeDriverManager().install()}
browser = Browser('chrome', **executable_path, headless=False)

# open browser with website
url='https://redplanetscience.com/'
browser.visit(url)
```

```python
html = browser.html
soup = bs(html, 'html.parser')

# find titles and summaries
titles = soup.find_all('div',class_='content_title')
summary = soup.find_all('div',class_='article_teaser_body')
```

```python
# import pandas dependency
import pandas as pd
```

```python
# create list for news 
mars_list = []

# for loop to find titles/summaries and store in list
for title in titles:
    mars_dict = {}
    mars_dict['title'] = title.text
    
    for summ in summary:
        mars_dict["preview"] = summ.text
        mars_list.append(mars_dict)
```
